```python
hsbc = yf.Ticker("0005.hk")

data = hsbc.history(period="1mo")

print(data)

opendata = np.array(data["Open"])
closedata = np.array(data["Close"])
plt.plot(opendata, label="open")
plt.legend()
plt.plot(closedata, label="close")
plt.legend()
plt.show
```

```python
hsbc = yf.Ticker("0005.hk")

data = hsbc.history(start="2020-01-01", end="2020-12-31")

hsbcdata = {
    "open": data["Open"],
    "close": data["Close"],
    "high": data["High"],
    "low": data["Low"]
}

hsbcdf = pd.DataFrame(hsbcdata)
print(hsbcdf)

json = hsbcdf.to_json(orient="index", date_format="iso")
print(json)
```
